```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x = pd.read_csv("xdata.csv")
y = pd.read_csv("ydata.csv")
X = x.values
X = X[:,1:]
x1 = x.iloc[:,1].values
x2 = x.iloc[:,2].values
y = y.values[:,1]
print(y.shape)
query = [2,3]
plt.scatter(query[0],query[1],c = 'r')
plt.scatter(x1,x2,c =y)
plt.show()

```

```python
def dist(x2,x1):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,query_point,k = 5):
    val = []
    num = X.shape[0]
    
    for i in range(num):
        
        d = dist(query_point,X[i])
        val.append((d,Y[i]))
    val.sort(key = lambda x: x[0])
    val = val[:k]
    vals = np.array(val)
    print(vals)
    new_val = np.unique(vals[:,1],return_counts = True)
    print(new_val)
    idx = np.argmax(new_val[1])
    pred = new_val[0][idx]
    return pred
    
pred = knn(X,y,query_point = [2,3],k=5)
print(pred)
```
